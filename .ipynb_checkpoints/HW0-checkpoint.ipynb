```python
# Monty hall problem
def simulate_prizedoor(nsim):
    
    # random array of 0s, 1s and 2s
    answer = np.random.randint(3,size=nsim)
    return answer
```

```python
prize_door = simulate_prizedoor(3)
```

```python
def simulate_guess(nsim):
    
    # random array of 0s, 1s and 2s
    answer_guess = np.random.randint(3,size=nsim)
    return answer_guess
```

```python
guess = simulate_guess(3)
```

```python
def goat_door(prizedoors, guesses):
    i=0
    goat_door = []
    for each in xrange(len(prizedoors)):
        if i < len(prizedoors) and i < len(guesses):
            if prizedoors[each] == 0 and guesses[each] == 0:
                goat_door.append(1)
            elif prizedoors[each] == 0 and guesses[each] == 1:
                goat_door.append(2)
            elif prizedoors[each] == 1 and guesses[each] == 0:
                goat_door.append(2)
            elif prizedoors[each] == 1 and guesses[each] == 1:
                goat_door.append(2)
            elif prizedoors[each] == 0 and guesses[each] == 2:
                goat_door.append(1)
            elif prizedoors[each] == 2 and guesses[each] == 0:
                goat_door.append(1)
            elif prizedoors[each] == 2 and guesses[each] == 2:
                goat_door.append(0)
            elif prizedoors[each] == 2 and guesses[each] == 1:
                goat_door.append(0)
            elif prizedoors[each] == 1 and guesses[each] == 2:
                goat_door.append(0)
            elif prizedoors[each] == 2 and guesses[each] == 2:
                goat_door.append(1)
            i += 1
        else:
            print "There are not enough options in the prizedoors or guessdoors"
    return goat_door
        
        
```

```python
print goat_door(np.array([0, 1, 2]), np.array([1, 1, 1]))
```

```python
def win_percentage(prizedoors, guesses):
    i=0
    count_match = 0
    count_donotmatch = 0
    for each in xrange(len(prizedoors)):
        if i < len(prizedoors) and i < len(guesses):
            if prizedoors[each] == guesses[each]:
                count_match += 1
            else:
                count_donotmatch +=1
            i += 1
    per_match = float(count_match)/len(prizedoors) * 100
    return per_match
```

```python
print win_percentage(np.array([0, 1, 2]), np.array([0, 1, 2]))
```

```python
def door_change(prizedoors, guesses):
    i=0
    new_door = []
    goat_doors = goat_door(prizedoors, guesses)
    for each in xrange(len(prizedoors)):
        if i < len(goat_doors) and i < len(guesses):
            if goat_doors[each] == 0 and guesses[each] == 0:
                new_door.append(1)
            elif goat_doors[each] == 0 and guesses[each] == 1:
                new_door.append(2)
            elif goat_doors[each] == 1 and guesses[each] == 0:
                new_door.append(2)
            elif goat_doors[each] == 1 and guesses[each] == 1:
                new_door.append(2)
            elif goat_doors[each] == 0 and guesses[each] == 2:
                new_door.append(1)
            elif goat_doors[each] == 2 and guesses[each] == 0:
                new_door.append(1)
            elif goat_doors[each] == 2 and guesses[each] == 2:
                new_door.append(0)
            elif goat_doors[each] == 2 and guesses[each] == 1:
                new_door.append(0)
            elif goat_doors[each] == 1 and guesses[each] == 2:
                new_door.append(0)
            elif goat_doors[each] == 2 and guesses[each] == 2:
                new_door.append(1)
            i += 1
        else:
            print "There are not enough options in the prizedoors or guessdoors"
    return new_door
    
```

```python
print door_change(np.array([0, 1, 2]), np.array([1, 1, 1]))
```

```python
#For 10000 simulations
prize_door = simulate_prizedoor(10000)
guess = simulate_guess(10000)
goat_door(prize_door, guess)
```

```python
#Win percentage where contestant keeps his original guess:
print win_percentage(prize_door, guess)
```

```python
#Win percentage where contestant keeps his original guess:
new_doors = door_change(prize_door, guess)
print win_percentage(prize_door, new_doors)
```
